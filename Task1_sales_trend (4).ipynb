```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/Retail and wherehouse Sale (1).csv')

# DATA CLEANING

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Remove missing values
df = df.dropna()

# Remove duplicate rows
df = df.drop_duplicates()

# Convert retail sales column to numeric
df['RETAIL SALES'] = pd.to_numeric(df['RETAIL SALES'], errors='coerce')

# Remove rows with invalid sales values
df = df.dropna(subset=['RETAIL SALES'])

# Display cleaned data
print(df.head())

# DATA VISUALIZATION

monthly_sales = df.groupby('MONTH')['RETAIL SALES'].sum()

monthly_sales.plot(kind='line', marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Retail Sales")

plt.grid(True)

plt.show()
```
